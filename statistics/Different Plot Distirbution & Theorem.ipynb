Write a Pandas program to calculate IQR and detect outliers.

````python
import pandas as pd

data = pd.Series([5, 7, 8, 9, 10, 12, 15, 18, 50])

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

outliers = data[(data < Q1 - 1.5*IQR) | (data > Q3 + 1.5*IQR)]
print(outliers)
````

## Task: Write a function what will filter non outliers data ?

````python
data = [5, 7, 8, 9, 10, 12, 15, 18, 50]
data
````

✅ Function: Filter Non-Outliers (IQR Method)

````python
def filter_non_outliers(data):
    """
    Removes outliers from the dataset using the IQR method.

    """
    ## Task1: Sort the data
    data = sorted(data) 
    ## Task2: Calculate Q1 and Q3
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1

    ## Task3: Determine lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    ## Task4: Filter the data
    return [x for x in data if lower_bound <= x <= upper_bound]

# Call the function with your dataset
clean_data = filter_non_outliers(data)
print(clean_data)
````

````python
# Basic

result = []

for x in data:
    if Lower_bound <= x <= Upper_bound:  # Is x >= lower_bound AND x <= upper_bound? If yes, keep x if no, discard x
        result.append(x)

print("Data without outliers:", result)
````
