# Matplotlib

```python
import matplotlib.pyplot as plt
```

📊 Basic Line Plot

```python
plt.plot([1,2,3], [10,20,30])
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

📈 Line Plot with Markers

```python
plt.plot([1,2,3], [10,20,30], marker='o')
plt.show()
```

📊 Bar Plot

```python
x = ['A','B','C']
y = [10,20,15]

plt.bar(x, y)
plt.title("Bar Chart")
plt.show()
```

📊 Horizontal Bar

```python
plt.barh(x, y)
plt.show()
```

🔵 Scatter Plot

```python
plt.scatter([1,2,3], [10,20,15])
plt.title("Scatter Plot")
plt.show()
```

📉 Histogram

```python
data = [10,20,20,30,30,30,40]

plt.hist(data, bins=5)
plt.title("Histogram")
plt.show()
```

📈 Multiple Lines

```python
x = [1,2,3]

plt.plot(x, [10,20,30], label='Line 1')
plt.plot(x, [15,25,35], label='Line 2')

plt.legend()
plt.show()
```

🎨 Grid + Style

```python
plt.plot([1,2,3], [10,20,30])
plt.grid(True)
plt.show()
```

🔄 Subplots

```python
fig, ax = plt.subplots(1,2)

ax[0].plot([1,2,3], [10,20,30])
ax[1].bar(['A','B'], [5,10])

plt.show()
```

🎯 Key Takeaways

```python
"""
✅ Use plt.plot() for line charts
✅ Use plt.scatter() for relationships
✅ Use plt.bar() for categorical data
✅ Always use plt.show() to display
✅ Add title, xlabel, ylabel for clarity
✅ Use legend() for multiple lines
"""
```
