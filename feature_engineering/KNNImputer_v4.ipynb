# KNN Imputer

* KNNImputer is much easier to understand than IterativeImputer if you think in terms of “similar rows help each other.”
* “Find rows that are most similar (neighbors), and use their values to fill missing data.”

| Method    | Idea                     |
| --------- | ------------------------ |
| KNN       | “Use similar rows”       |
| Iterative | “Build model to predict” |

```bash
KNNImputer fills missing values by looking at similar data points and averaging their values.

````python
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame({
    'age':[25, 30, np.nan, 40, 35],
    'salary':[50000, 60000, 52000, 80000, 75000]
})
print(df)
````

👉 Missing value: age at index 2

🔍 Step-by-Step What Happens

```bash
We want to fill:

age = ?
salary = 52000

Step 1: Find Similar Rows (Neighbors)

We compare row 2 with others using salary (since age is missing).

| row | salary | distance |
| --- | ------ | -------- |
| 0   | 50000  | 2000     |
| 1   | 60000  | 8000     |
| 3   | 80000  | 28000    |
| 4   | 75000  | 23000    |

```bash
Missing_age at index 2, Salary 52000
So:
Index  Calcuation     Distance
0     52000 - 50000    2000
1     60000 - 52000    8000
3     80000 - 52000    28000 & So on!

👉 Closest rows = row 0 and row 1 : Where less Distance Like 2000 & 8000

Step 2: Pick K neighbors

```bash
KNNImputer(n_neighbors=2)

👉 Neighbors = row 0 and row 1

Step 3: Take their ages
| row | age |
| --- | --- |
| 0   | 25  |
| 1   | 30  |


Step 4: Average them

```bash
age = (25 + 30) / 2 = 27.5


```bash

✅ Final Result

| index | age  | salary |
| ----- | ---- | ------ |
| 2     | 27.5 | 52000  |


🧪 Apply in Code

````python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2)

df_imputed = imputer.fit_transform(df)

df_imputed = pd.DataFrame(df_imputed, columns=df.columns)
print(df_imputed)
````

🔥 Intuition (Super Simple)

```bash
“Find people with similar salary → use their age”

⚠️ Important Detail: Distance

```bash
It uses Euclidean distance:
distance = √((x1 - x2)² + (y1 - y2)²)

👉 But only uses available (non-missing) features

```bash

🧠 When to Use KNNImputer
✅ Good when:
Data has patterns
Similar rows exist
Small/medium dataset

❌ Not ideal when:
Very large dataset (slow)
Features not meaningful
Too many missing values
