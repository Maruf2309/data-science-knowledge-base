3. SQL vs Pandas (Super Important)

```bash

SELECT *
FROM (
    SELECT *,
           DENSE_RANK() OVER (PARTITION BY Department ORDER BY Salary DESC) as rnk
    FROM table
) t
WHERE rnk <= 3;

💡 Pro Insight (Interview Gold ⭐)

```bash

| Concept        | Pandas            | SQL                |
| -------------- | ----------------- | ------------------ |
| Grouping       | `groupby()`       | `PARTITION BY`     |
| Ranking        | `rank()`          | `DENSE_RANK()`     |
| Filtering rank | boolean filtering | `WHERE` / subquery |


🎯 Problem

Find employees whose salary is greater than their department average

````python
# Step 1: Calculate department average

df['dept_avg_salary'] = df.groupby('Department')['Salary'].transform('mean')  # 👉 transform('mean') returns the average for each row’s department

# 🔹 Step 2: Filter employees above average
result = df[df['Salary'] > df['dept_avg_salary']]

# 📊 Final Code (Clean Version)
df['dept_avg_salary'] = df.groupby('Department')['Salary'].transform('mean')
result = df[df['Salary'] > df['dept_avg_salary']]
result

````

```bash
🔄 SQL Equivalent
SELECT *
FROM table t
WHERE Salary > (
    SELECT AVG(Salary)
    FROM table
    WHERE Department = t.Department
);
